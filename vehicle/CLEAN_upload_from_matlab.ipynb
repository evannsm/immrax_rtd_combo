```python
from scipy.io import loadmat
import numpy as np  
import matplotlib.pyplot as plt 
```

```python
# robot desired location
x_des = 0.75
y_des = 0.5
```

```python
# Control Data
data = loadmat('export_for_verification/data1.mat')
T_des = data['T_brk'][0]
U_des = data['U_brk']
Z_des = data['Z_brk']


# Obstacle Data
obstacle = loadmat('export_for_verification/obs1.mat')
O = obstacle['O']
O_buf = obstacle['O_buf']
O_pts = obstacle['O_pts']
```

```python
plt.figure(1) # Create a Figure
# plt.subplot(1, 2, 1) # Create subplot with 1 row, 3 columns, and select the third subplot
plt.axis('equal') # Set aspect ratio to be equal
plt.rc('font', size=15) # Adjust font size
plt.fill(O_buf[0], O_buf[1], color=[1, 0.5, 0.6]) # Plot the buffered obstacle (assuming O_buf contains the data for the obstacle)
plt.fill(O[0], O[1], color=[1, 0.7, 0.8]) # plot actual obstacle
plt.plot(O_pts[0,:],O_pts[1,:], '.', color=[0.5, 0.1, 0.1], markersize=15) # plot discretized obstacle
plt.plot(x_des, y_des, 'k*', linewidth=2, markersize=15) # Plot the desired location of the robot
plt.show() # Show the plot
```
